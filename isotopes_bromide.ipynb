### Bromide

```python
#the measured concentrations are taken from a dilution of the samples in 150ml water
#the concentration has to be referred to the original volume
def Br_c2m(cBr,V_dilute=0.15,V_sample='quicksampler',soil_conc=False):
    #V_dilute=0.15 #standard procedure with 0.15 litre
    #cBr concentration in samples in mg/l
    if V_sample=='quicksampler':
        V_soilsample=(np.pi*0.021**2 * 0.048)*1000. # sample volume 0.0211 litre
    else:
        V_soilsample=V_sample

    mBr=cBr*V_dilute
    if soil_conc:
        #also calculate concentration in soil sample
        csBr=mBr/V_soilsample
        return [mBr,cBr]
    else:
        return mBr
```

```python
#recovery rate calculation
def Br_recov(bb,bb2=False,irr=50.,cKBr=5.0264,coredepth=1.):
    #irr=50. #l irrigation
    #cKBr=5.0264 #g/l KBr
    #coredepth=1. #m depth of drilled core
    cBr_minus=cKBr*(79.904/(79.904+39.0983)) #reduce KBr to Br- ions only
    mBr_minus=cBr_minus*irr*1000. #irrigared Br- in mg
    
    #sampled profile:
    mBr_recovered=Br_c2m(bb).sum().sum() #recovered Br- in mg
    
    V_soilsample=(np.pi*0.021**2 * 0.048)
    V_sampled=bb.shape[0]*bb.shape[1]*V_soilsample #sampled volume in profile

    irr_area=1. #1m2 sprinkled area
    z_sampled=bb.shape[0]*0.05 #depth of sampled profile
    
    if any(bb2)==False:
        scaling=(z_sampled*irr_area)/V_sampled #reference sampled volume to soil volume
        mBr_scaled=mBr_recovered*scaling
        recovery=mBr_scaled/mBr_minus
        return recovery
    else:
        #drilled core
        mBr_core=Br_c2m(bb2,0.1,np.pi*0.04**2 * 0.05).sum() #recovered Br- in mg (assuming 80mm drill and dilution with 100ml)
        V_core=coredepth*np.pi*0.04**2
        scaling2=(z_sampled*irr_area)/(V_sampled+V_core)
        mBr_scaled2=(mBr_recovered+mBr_core)*scaling2
        recovery2=mBr_scaled2/mBr_minus
        return recovery2
```

```python
def plotbrprofile(InfPlot,savef=False,xlims=(0.0,70.0),ylims=(-1.0,0.0),fsiz=(5,3),imx=5,irr=50.,cKBr=5.0264,coredepth=1.):
    if InfPlot=='X':
        InfID='Inf1'
    elif InfPlot=='XI':
        InfID='Inf2'
    elif InfPlot=='XII':
        InfID='Inf3'
    else:
        InfID=False

    con=BR.c_Br[BR.Plot==InfPlot]
    row=BR.row[BR.Plot==InfPlot].fillna(0)
    col=BR.col[BR.Plot==InfPlot].fillna(0)
    col[col>10]=0

    img = np.zeros((row.max(), col.max()))
    img[row.astype(np.int)-1,col.astype(np.int)-1] = con

    bb=pd.DataFrame(img.T)
    bb.columns=((bb.columns+0.5)*-0.05)
    bb=bb.T

    if InfID:
        bb2=BRc.c_Br[BRc.Plot==InfID]
        bb2.index=np.round(BRc.z[BRc.Plot==InfID]*((100.*coredepth)/BRc.z[BRc.Plot==InfID].min())*-0.01,3)
    
        #recovery rate
        recovRate=Br_recov(bb,bb2,irr,cKBr,coredepth)

    recovRate_profile=Br_recov(bb,False,irr,cKBr,coredepth)
    
    #plotting
    fig, ax1 = plt.subplots(figsize=fsiz)
    plt.subplot(121)
    plt.plot(Br_c2m(bb).mean(axis=1),bb.index,alpha=1,lw=2,label='profile')#color='#20B2AA'
    if InfID:
        plt.plot(Br_c2m(bb2,0.1,np.pi*0.04**2 * 0.05).values,bb2.index,alpha=1,lw=2,color='#FF6347',label='core')
    plt.fill_betweenx(bb.index.values.astype('float64'),Br_c2m(bb).min(axis=1).values, Br_c2m(bb).max(axis=1).values,alpha=0.3)#,color=colr)
    
    plt.xlim(xlims)
    plt.ylim(ylims)
    if InfID:    
        plt.legend(loc=4)
        recov_text=''.join(['R$_{profile}$=',str(np.round(recovRate_profile,2)),'\nR$_{all}$=',str(np.round(recovRate,2))])
        plt.text(xlims[1],ylims[1],recov_text,horizontalalignment='right', verticalalignment='top')
    else:
        recov_text=''.join(['R$_{recov}$=',str(np.round(recovRate_profile,2))])
        plt.text(xlims[1],ylims[0],recov_text,horizontalalignment='right', verticalalignment='bottom')
    plt.title(''.join(['Bromide Recovery \nPlot ',InfPlot]),loc='left')
    plt.ylabel('Depth [m]')
    plt.xlabel('m(Br-) [mg]')
    
    plt.grid(b=True,linestyle='--', which='major', color='lightgrey',alpha=0.3)
    #ax1.set_axisbelow(True)
    plt.tight_layout()
    sns.despine(offset=5)
        
    plt.subplot(122)
    imshow(Br_c2m(img), interpolation='nearest', extent=(0,imx,0,int(ylims[0]*-100./5)-1),vmin=0.0, vmax=xlims[1], cmap='GnBu')
    #title("Bromid Profile")
    #xlabel('column [a 5cm]')
    sns.despine(offset=5)
    yticks([])
    #ylabel('row [a 5cm]')
    c1=colorbar()
    #c1.set_label('c(Br-) [mg/l]')

    plt.show()
    if savef:
        fig.savefig(savef)
```
